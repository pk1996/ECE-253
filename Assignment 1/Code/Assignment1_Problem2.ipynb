Import libraries

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

2)\
(i)Read image

```python
A = cv2.imread('/home/parth/work/UCSD/Fall 2021/ECE 253 Image Processing/Assignment 1/geisel.jpg')
A = cv2.cvtColor(A, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(A)
plt.show()
```

2)\
(ii)Transform into grayscale

```python
B = cv2.cvtColor(A, cv2.COLOR_RGB2GRAY)
print('Pixel intensities range ', np.min(B), np.max(B))
plt.axis('off')
plt.imshow(B, cmap = 'Greys')
plt.show()
```

2)\
(iii)Add 15 to each pixel value

```python
C = B + 15
C[C>255] = 255
plt.imshow(C, cmap = 'Greys')
plt.axis('off')
plt.show()
```

2)\
(iv)Flip along horizontal and vertical axis

```python
D = np.flip(B, axis = 1) # Horizontal Flip
D = np.flip(D, axis = 0) # Vertically Flip
plt.imshow(D, cmap = 'Greys')
plt.axis('off')
plt.show()
```

2)\
(v)Thresholding based on median filter

```python
medianB = np.median(B)
print('Median of all values in B is',medianB)
E = np.zeros_like(B)
E[B<=medianB] = 1;
plt.imshow(E, cmap = 'Greys')
plt.axis('off')
plt.show()
```
